# Analysis of Super Store

# Load and Drop Unwanted Data

```python
# Step -1- Import the Necessary Libraries like pandas -> frame the dataset, matplotlib -> visualization the data
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step -2- Load the Dataset and show it.
data_frame=pd.read_csv(r"C:\Users\OM MURUGA\Downloads\Analysis of Super Store - DA.csv")
data_frame
```

```python
# Step -3- Drop the Unwanted Columns in Dataset like postal code and country because the country is unique.
data_frame=data_frame.drop("Postal Code",axis="columns")
data_frame=data_frame.drop("Country",axis="columns")
data_frame
```

```python
# Step -4- Find Datatype of the dataset that is helpful for visualization.
data_frame.info()
```

# Visualize and Analysis the data using graphs.

```python
# Step -5- State vs Sales using barchart
data_frame.groupby("State")["Sales"].sum().plot.bar()
```

```python
# Step -6- State vs Profit using barchart
data_frame.groupby("State")["Profit"].sum().plot.bar()
```

# From the above two graphs we see that the states like Texas and Pennsylvania gives the negative profit but the sales are high.
# To find the reason for that.

```python
# Step -7- Region vs Sales using piechart
data_frame.groupby("Region")["Sales"].sum().plot.pie(autopct="%.2f")
```

```python
# Step -8- Region vs Sales using piechart
data_frame.groupby("Region")["Profit"].sum().plot.pie(autopct="%.2f")
```

# From the above two graphs we see that,
# Sales  : WEST > EAST > CENTRAL > SOUTH
# Profit : WEST > EAST > SOUTH > CENTRAL
# we see that the Sales is larger in Central region but the profit is very low.
# To find the reason for that.

```python
# Step -9- Region vs Discount using piechart
data_frame.groupby("Region")["Discount"].sum().plot.pie(autopct="%.2f")
```

# Conclusion :

# From the above graph we see that,
#     Central Region has more discounts than other Regions so, that the profit in the states like Texas(central) and Illinois(central) are decreases as the result the profit in the Central Region are also Decreases.
